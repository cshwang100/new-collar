```python
apt=[]
for i in range(1,7):
    url=f"https://hogangnono.com/api/v2/offers?page={i}&regionCode=11&x=0&y=0"
    res=requests.get(url, verify=False)
    data_parse=json.loads(res.text)
    for value in data_parse["data"]["data"]:
        data=[]
        data.append(value["apt"]["name"])
        data.append(value["apt"]["hash"])
        data.append(value["apt"]["regionName"])
        data.append(value["dateText"])
        data.append(value["dDayText"])           
        data.append(value["apt"]["viewCount"])
        data.append(value["apt"]["offerSubscriptionCount"])
        apt.append(data)              
```

```python
apt_서울=pd.DataFrame(apt, columns=["이름", "아파트id","분양지역", "분양예정일1", "분양예정일2","방문객수","알람신청수"])
```
